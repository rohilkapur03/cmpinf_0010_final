##### In this first data we loaded the Land Use Areas with geopandas.

```python
import pandas as pd 

#Load Trees
trees_data = pd.read_csv('trees.csv', low_memory=False)
print(trees_data.head(5))
```

##### In this data, we see the amount of trees that are planted in an area. The latitude and longitude will establish which trees are in which city. The city with the most trees will get more points, and thus be better based on this score.

```python
import pandas as pd

# Create a new empty DataFrame called 'trees'
trees = pd.DataFrame(columns=["Location", "Tree Name", "Lat", "Long", "Neighborhood", "Count"])

# Set the 'Tree Name', 'Lat', 'Long', and 'Neighborhood' columns in 'trees' based on values from 'data'
trees["Tree Name"] = trees_data["common_name"]
trees["Lat"] = trees_data["latitude"]
trees["Long"] = trees_data["longitude"]
trees["Neighborhood"] = trees_data["neighborhood"]

# Set the 'Count' column to zero for all rows
trees["Count"] = 0

# Show the first five rows of the 'trees' DataFrame
trees.head(5)
```

##### Now, we need to check how many trees are in each neighborhood. 

```python
import pandas as pd

# Load the CSV file into a pandas dataframe
df = pd.read_csv('trees.csv', low_memory=False)

# Extract the Neighborhood column
neighborhoods = df['neighborhood']

# Count the number of unique neighborhoods
num_unique_neighborhoods = len(neighborhoods.unique())

# Print the result
print('There are', num_unique_neighborhoods, 'unique neighborhoods in the file.')
```

```python
tree_counts = trees.groupby("Neighborhood").size().reset_index(name="Tree Count")
print(tree_counts)
```

Now, to sort the data from the highest Tree Count to the lowest. 

```python
tree_counts_sorted = tree_counts.sort_values("Tree Count", ascending=False)
print(tree_counts_sorted)
```

Now, to make a graph showing this data with the top 30 count. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the trees.csv file into a pandas dataframe
trees = pd.read_csv('trees.csv')

# Group the trees by neighborhood and count the number of trees in each neighborhood
tree_counts = trees.groupby('neighborhood')['id'].count().reset_index()

# Rename the 'id' column to 'Tree Count'
tree_counts.rename(columns={'id': 'Tree Count'}, inplace=True)

# Sort the tree counts in descending order
tree_counts_sorted = tree_counts.sort_values('Tree Count', ascending=False)

# Plot the top 30 neighborhoods with the most trees
plt = tree_counts_sorted.iloc[:30].plot(kind='bar', x="neighborhood", y="Tree Count", title="Tree Count in Pittsburgh")
plt.set_xlabel("Neighborhood")
plt.set_ylabel("Tree Count")
plt.show()
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# Read the trees.csv file into a pandas dataframe
trees = pd.read_csv('trees.csv')

# Create a new folium map centered at Pittsburgh
map = folium.Map(location=[40.4406, -79.9959], zoom_start=12)

# Group the trees by latitude and longitude and count the number of trees in each location
tree_counts = trees.groupby(['latitude', 'longitude'])['id'].count().reset_index()

# Create a heatmap layer based on the tree counts
heatmap_layer = HeatMap(data=tree_counts[['latitude', 'longitude', 'id']].values.tolist(), 
                        name='Tree Count', 
                        min_opacity=0.3, 
                        max_val=float(tree_counts['id'].max()), 
                        radius=10, 
                        blur=15, 
                        max_zoom=1)

# Add the heatmap layer to the folium map
heatmap_layer.add_to(map)

# Display the map
map
```

### Metric 2: Schools within 5km
#### Analyzed by Rohil Kapur

First we need to import the neigborhoods into a dataframe using Pandas

```python
import pandas as pd

#load neighborhoods into CSV file 
neighborhoods_df = pd.read_csv('neighborhoods.csv')
neighborhoods_data = neighborhoods_df[['hood', 'intptlat10', 'intptlon10']]

neighborhoods_data.head(10)
```

Analyze the data and drop the rows that do not contain latitudes and longitudes

```python
neighborhoods_data.drop(index=15, inplace=True)

neighborhoods_data.head(20)
```

Now import the educational institutions into another dataframe with their neccessary columns to perform analysis.

```python
import pandas as pd

education_df = pd.read_csv('education.csv')
education_data = education_df[['name', 'address', 'latitude', 'longitude']]

education_data.head(20)
```

Now using geodesic functions, calculate the distance between each school and the list of neighborhoods. 

Then create a new dataframe to display the list of all neighborhoods and the number of schools within 5km.

```python
from geopy.distance import geodesic

# Define a function to calculate the distance between a neighborhood and a school
def distance(neighborhood_lat, neighborhood_lon, school_lat, school_lon):
    return geodesic((neighborhood_lat, neighborhood_lon),(school_lat, school_lon)).km

# Create an empty dictionary to store the number of schools within 5 km of each neighborhood
schools_within_5km = {}

# Loop over all neighborhoods
for index, neighborhood in neighborhoods_data.iterrows():
    neighborhood_name = neighborhood['hood']
    neighborhood_lat = neighborhood['intptlat10']
    neighborhood_lon = neighborhood['intptlon10']
    schools_within_5km_count = 0
    # Loop over all schools
    for index, school in education_data.iterrows():
        school_lat = school['latitude']
        school_lon = school['longitude']
        school_distance = distance(neighborhood_lat, neighborhood_lon, school_lat, school_lon)
        # Only consider schools that are within 5 km of the neighborhood
        if school_distance <= 5:
            schools_within_5km_count += 1
    schools_within_5km[neighborhood_name] = schools_within_5km_count

# Create a DataFrame with the number of schools within 5 km of each neighborhood
schools_within_5km_df = pd.DataFrame.from_dict(schools_within_5km, orient='index', columns=['schools_within_5km'])
schools_within_5km_df.index.name = 'neighborhood'
schools_within_5km_df.reset_index(inplace=True)

# Print the resulting DataFrame
print(schools_within_5km_df.head(90))
```

Organize the dataframe based on decending order to find the top 10 neighborhoods with most schools within 5km. 

```python
schools_within_5km_df = schools_within_5km_df.sort_values(by='schools_within_5km', ascending=False)

schools_within_5km_df.head(10)
```

Create a plot to display the top 10 neighborhoods visually.

```python
import matplotlib.pyplot as plt

# Get the top 10 neighborhoods with the most schools within 5km
top_10_neighborhoods = schools_within_5km_df.sort_values(by='schools_within_5km', ascending=False).head(10)

# Create a bar plot of the neighborhoods and their school counts
plt.bar(top_10_neighborhoods['neighborhood'], top_10_neighborhoods['schools_within_5km'])
plt.xticks(rotation=90)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Schools within 5km')
plt.title('Top 10 Neighborhoods with the most Schools within 5km')
plt.show()
```

### Metric 3: Traffic Data
#### Analyzed by Ayan Shah

First import the traffic data from all the neighborhoods in Pittsburgh

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
traffic_data = pd.read_csv("traffic.csv")

traffic_data.head()
df = traffic_data[['neighborhood']]

traffic_data['neighborhood'].value_counts().loc[lambda x : x>10].plot(kind='bar', figsize=(15,5))
```

The data shows the distribution of camera between neighborhoods, which could cause a skew in the data as some neighborhoods. However, we will take that into account later.

The next data point will show the amount of bikers in a certain area. If there are a high number of bikers in a certain area, that would imply that area is safe.

```python
df = pd.read_csv("traffic.csv")
df = df[df['average_daily_bike_traffic'].notna()]
df.plot.bar(x='neighborhood', y='average_daily_bike_traffic', rot=0, figsize=(15,5))
```

This is giving us more information about the individual neighborhoods. There aren't a lot of individual datapoints for bike traffic, which is one of the drawbacks from this dataset, so we'll look at car traffic to determine safety in each neighborhood

```python
speeding = pd.read_csv('traffic.csv', usecols=["average_daily_car_traffic", "neighborhood"])
speeding = speeding[speeding['average_daily_car_traffic'].notna()]
speeding = speeding.groupby(['neighborhood']).sum()
speeding.drop(speeding[speeding['average_daily_car_traffic'] <= 50].index, inplace = True)
speeding.plot.bar(figsize=(15,5))
```

This is much more usable information. There are several neighborhoods that have very very high levels of traffic. This indicates that there are more cars in general in that area and would be generally an area to avoid as a door dash delivery driver. I would like to point out a few notable neighborhoods missing from the above group: the central business district, any oakland districts, and the south side flats. All neighborhoods which we have as top canidates for best neighborhood. One thing to consider is the lack of significant numbers in the central business district, which I think is explained by the very few datapoints we have in that area.

The final piece of information we are going to look at is speeding percent in each neighborhood.

```python
speeding = pd.read_csv('traffic.csv', usecols=["percent_over_limit", "neighborhood"])
speeding = speeding[speeding['percent_over_limit'].notna()]
speeding = speeding.groupby(['neighborhood']).mean()
speeding.drop(speeding[speeding['percent_over_limit'] <= 50].index, inplace = True)
speeding.plot.bar(figsize=(15,5))
```
