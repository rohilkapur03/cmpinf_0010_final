```python
import pandas as pd

#load neighborhoods into CSV file 
neighborhoods_df = pd.read_csv('neighborhoods.csv')
neighborhoods_data = neighborhoods_df[['hood', 'intptlat10', 'intptlon10']]

neighborhoods_data.head(10)
```

```python
neighborhoods_data.drop(index=15, inplace=True)

neighborhoods_data.head(20)
```

import the schools and education centers in the pittsburgh area 

```python
import pandas as pd

education_df = pd.read_csv('education.csv')
education_data = education_df[['name', 'address', 'latitude', 'longitude']]

education_data.head(20)
```

use geodesic functions to calculate the distances to each school

```python
from geopy.distance import geodesic

# Define a function to calculate the distance between a neighborhood and a school
def distance(neighborhood_lat, neighborhood_lon, school_lat, school_lon):
    return geodesic((neighborhood_lat, neighborhood_lon),(school_lat, school_lon)).km

# Create an empty dictionary to store the number of schools within 5 km of each neighborhood
schools_within_5km = {}

# Loop over all neighborhoods
for index, neighborhood in neighborhoods_data.iterrows():
    neighborhood_name = neighborhood['hood']
    neighborhood_lat = neighborhood['intptlat10']
    neighborhood_lon = neighborhood['intptlon10']
    schools_within_5km_count = 0
    # Loop over all schools
    for index, school in education_data.iterrows():
        school_lat = school['latitude']
        school_lon = school['longitude']
        school_distance = distance(neighborhood_lat, neighborhood_lon, school_lat, school_lon)
        # Only consider schools that are within 5 km of the neighborhood
        if school_distance <= 5:
            schools_within_5km_count += 1
    schools_within_5km[neighborhood_name] = schools_within_5km_count

# Create a DataFrame with the number of schools within 5 km of each neighborhood
schools_within_5km_df = pd.DataFrame.from_dict(schools_within_5km, orient='index', columns=['schools_within_5km'])
schools_within_5km_df.index.name = 'neighborhood'
schools_within_5km_df.reset_index(inplace=True)

# Print the resulting DataFrame
print(schools_within_5km_df.head(90))
```

## sort the neighborhoods in numerical order with decreasing to see which neighborhood has most schools

```python
schools_within_5km_df = schools_within_5km_df.sort_values(by='schools_within_5km', ascending=False)

schools_within_5km_df.head(91)
```

now we will create a bar graph to display the best 10 neighborhoods

```python
import matplotlib.pyplot as plt

# Get the top 10 neighborhoods with the most schools within 5km
top_10_neighborhoods = schools_within_5km_df.sort_values(by='schools_within_5km', ascending=False).head(10)

# Create a bar plot of the neighborhoods and their school counts
plt.bar(top_10_neighborhoods['neighborhood'], top_10_neighborhoods['schools_within_5km'])
plt.xticks(rotation=90)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Schools within 5km')
plt.title('Top 10 Neighborhoods with the most Schools within 5km')
plt.show()
```
